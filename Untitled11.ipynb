```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to uploaded CSV
csv_path = "sample_data.csv"

# Step 1: Load CSV
df = pd.read_csv(csv_path)

# Step 2: Clean Data
df.dropna(how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)
df.columns = [col.strip().replace(" ", "_").lower() for col in df.columns]
df.fillna(method='ffill', inplace=True)

# Try converting numeric columns
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

# Step 3: Summary Statistics
print("📊 Summary Statistics:\n")
print(df.describe(include='all'))

# Step 4: Sample Grouping (depends on your data)
if 'year' in df.columns and 'value' in df.columns:
    df_grouped = df.groupby('year')['value'].sum().reset_index()

# Step 5: Visualizations

# 1. Line Plot: Value over Time
if 'year' in df.columns and 'value' in df.columns:
    plt.figure()
    sns.lineplot(data=df_grouped, x='year', y='value')
    plt.title("Trend Over Time")
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()

# 2. Bar Plot: Total Value by Category
if 'category' in df.columns and 'value' in df.columns:
    df_cat = df.groupby('category')['value'].sum().sort_values()
    plt.figure()
    df_cat.plot(kind='barh')
    plt.title("Total Value by Category")
    plt.xlabel("Value")
    plt.ylabel("Category")
    plt.grid(True)
    plt.show()

# 3. Box Plot: Value Distribution by Category
if 'category' in df.columns and 'value' in df.columns:
    plt.figure()
    sns.boxplot(x='category', y='value', data=df)
    plt.title("Value Distribution by Category")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Step 6: Export Cleaned Data
df.to_csv("cleaned_sample_data.csv", index=False)

# Step 7: Summary
print("\n🔍 FINDINGS SUMMARY:")
print("1. Trend shows how values evolve over time.")
print("2. Category-wise totals reveal dominant segments.")
print("3. Some categories have wide value ranges, suggesting variation.")



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Average salary by department
plt.figure(figsize=(8,5))
sns.barplot(x='department', y='salary', data=df, estimator='mean')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()

# 3. Age vs Salary scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='salary', hue='department', data=df)
plt.title('Salary vs Age by Department')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(title='Department')
plt.show()
```
